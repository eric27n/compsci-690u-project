```python
from Bio import SeqIO
# Needed to create fasta db:
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import json
```

```python
from datasets import load_dataset

ds = load_dataset("tattabio/mibig_classification_dna")
```

```python
ds['test'][0]
```

```python
# Converting mibig dataset into fasta (for blast)
mibig_train = []
mibig_test = []
train_id_to_class = {} # fasta id -> correct class
test_id_to_class = {} # fasta id -> correct class

for instance in ds['train']:
    seq = Seq(instance['Sequence'])
    entry = instance['Entry']
    record = SeqRecord(Seq(seq), id=entry, description= '')
    mibig_train.append(record)
    
    train_id_to_class[entry] = instance['simple_class']
    
for instance in ds['test']:
    seq = Seq(instance['Sequence'])
    entry = instance['Entry']
    record = SeqRecord(Seq(seq), id=entry, description= '')
    mibig_test.append(record)
    
    test_id_to_class[entry] = instance['simple_class']
    
train_filename = 'mibig_train.fasta'
test_filename = 'mibig_test.fasta'

# Save fasta files
SeqIO.write(mibig_train, train_filename, 'fasta')
SeqIO.write(mibig_test, test_filename, 'fasta')

# Save json files
with open("train_classes.json", "w") as f:
    json.dump(train_id_to_class, f)
with open("test_classes.json", "w") as f:
    json.dump(test_id_to_class, f)
```
